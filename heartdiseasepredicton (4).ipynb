```python
# Putting feature variable to X
X = df.drop('target',axis=1)
# Putting response variable to y
y = df['target']
```

```python
# now lets split the data into train and test
from sklearn.model_selection import train_test_split

# Splitting the data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
X_train.shape, X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=42, n_jobs=-1, max_depth=5,
                                       n_estimators=100, oob_score=True)

model.fit(X_train, y_train)
```

```python
model.oob_score_
```

```python
rf = RandomForestClassifier(random_state=42, n_jobs=-1)

params = {
    'max_depth': [2,3,5,10,20],
    'min_samples_leaf': [5,10,20,50,100,200],
    'n_estimators': [10,25,30,50,100,200]
}

from sklearn.model_selection import GridSearchCV

# Instantiate the grid search model
grid_search = GridSearchCV(estimator=rf,
                           param_grid=params,
                           cv = 4,
                           n_jobs=-1, verbose=1, scoring="accuracy")

grid_search.fit(X_train, y_train)
```

```python
grid_search.best_score_
```

```python
rf_best = grid_search.best_estimator_
rf_best
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(30, 15))  # Prdor dimensione m t vogla
plot_tree(rf_best.estimators_[5], 
          feature_names=X.columns, 
          class_names=['Disease', "No Disease"], 
          filled=True)

# Shtimi i titullit
plt.title("Vizualizimi i Pems s Vendosjes nga Random Forest")

# Shfaqja e figurs
plt.show()
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(20,10))
plot_tree(rf_best.estimators_[7], feature_names = X.columns,class_names=['1', "0"],filled=True);
```
